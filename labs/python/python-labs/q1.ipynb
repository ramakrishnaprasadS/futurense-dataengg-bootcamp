```python
import mysql.connector as mycon
import pandas as pd
from datetime import datetime
hostname='localhost'
database="MoviesDB"
username="root"
pwd="password"
port_id=3308
```

```python

print("Select your choice 1.Add new movie\n 2.view movie details by movie id\n 3.update movie details by movieid\n 4.delete movie  by movieid\n 5.list all movies")
choice=int(input())
if choice == 1:
    movie_id=input("enter movie_id")
    title=input("enter title")
    year_of_release=input("enter year_of_release")
    genre=input("genre")

    try:
        conn=mycon.connect(
        host=hostname,
        database=db,
        user=username,
        password=pwd,
        port=port_id)

        cur=conn.cursor()
        movies_insert_script='INSERT INTO movies(movie_id,title,year_of_release,genre,create_date_time,update_timestamp) VALUES(%s,%s,%s,%s)'
        movies_insert_values=[movie_id,title,year_of_release,genre,str(datetime.now()),str(datetime.now())]
        cur.execute(movies_insert_script,movies_insert_values)         
        conn.commit()

    except Exception as error:
        print(error)

    finally:
        if cur is not None:
            cur.close()
        if conn is not None:
            conn.close()

elif choice == 3:
    movie_id=input("enter movie_id to be updated")
    title=input("enter title")
    year_of_release=input("enter year_of_release")
    genre=input("genre")

    try:
        conn=mycon.connect(
        host=hostname,
        database=db,
        user=username,
        password=pwd,
        port=port_id)

        cur=conn.cursor()
        cur.execute('update movies set title={title},set year_of_release={year_of_release},genre={genre} where movie_id={movie_id}')                 
           
        conn.commit()

    except Exception as error:
        print(error)

    finally:
        if cur is not None:
            cur.close()
        if conn is not None:
            conn.close()

elif choice == 4:
    movie_id=input("enter movie_id to be deleted")
    
    try:
        conn=mycon.connect(
        host=hostname,
        database=db,
        user=username,
        password=pwd,
        port=port_id)

        cur=conn.cursor()
        cur.execute('DELETE FROM movies where movie_id={movie_id}')                  
           
        conn.commit()

    except Exception as error:
        print(error)

    finally:
        if cur is not None:
            cur.close()
        if conn is not None:
            conn.close()
elif choice == 5:
    
    try:
        conn=mycon.connect(
        host=hostname,
        database=db,
        user=username,
        password=pwd,
        port=port_id)

        cur=conn.cursor()
        cur.execute('select * from movies') 
        result=cur.fetchall()
        for eachrecord in result:
            print(eachrecord)                 
           
        conn.commit()

    except Exception as error:
        print(error)

    finally:
        if cur is not None:
            cur.close()
        if conn is not None:
            conn.close()
```
